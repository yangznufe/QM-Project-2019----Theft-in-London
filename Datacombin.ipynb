```python
# -*- coding: utf-8 -*-  
import os  
  
def listdir(path):  #传入存储的list
    lst = []
    for file in os.listdir(path):  
        file_path = os.path.join(path, file)  
        lst.append(file_path) 
    return lst
lst1 = listdir(r'N:\crime data\crimedata' )

lst = []
for d in lst1:
    lst = lst + listdir(d)
lst
```

# 整合下载自OD的数据
1. 重要建筑

```python
import pandas as pd
import numpy as np
```

```python
df_losa = pd.read_csv('N:\crime data\lsoa_data_0102.csv')
```

```python
df_losa
```

```python
imp_b = pd.read_csv('N:\crime data\infrastrite\important_building_data.txt')
imp_b
```

```python
df_combined_1 = df_losa
for i in pd.unique(imp_b['buildgthem']):
    ls = imp_b[imp_b['buildgthem'] == i].groupby('lsoa11cd').count()
    lst = ls[['id']]
    lst['LSOA11CD'] = lst.index
    l = lst.rename(columns={'id': i})
    print(l.count())
    df_combined_1 = df_combined_1.merge(l, on = 'LSOA11CD',how = 'left' )
    
df_combined_1.count()
```

```python
lsoa_data_combined2 = df_combined_1.drop(columns = ['Emergency Service', 
       'Sports Or Exercise Facility', 'Road Transport', 'Retail', ' ',
       'Air Transport', 'Attraction And Leisure', 'Water Transport'])
```

```python
lsoa_data_combined2.to_csv('N:\crime data\lsoa_data_0102_2.csv')
```
