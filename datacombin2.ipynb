# Overlay Data to LSOA

1. building

```python
import pandas as pd
import numpy as np
import scipy as sp
from matplotlib import pyplot
```

```python
df_building = pd.read_csv(r'E:\Crime Porject\crime data\building\building_id_area.txt')
```

```python
df_building.head()
```

```python
df_building.columns
```

```python
g_build = df_building.groupby('LSOA11CD')
count = g_build[['ID']].count()
sum_g = g_build[['area']].sum() # m^2
count['LSOAID'] = count.index
sum_g['LSOAID'] = sum_g.index
lsoa_building_data = pd.merge(count,
                 sum_g,
                 on='LSOAID')
```

```python
lsoa_building = lsoa_building_data.rename(columns={"ID": "building_num", "area": "building_area"})
lsoa_building
```
